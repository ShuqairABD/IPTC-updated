```python
from iptcinfo3 import IPTCInfo
```

```python
from ipywidgets import interact_manual, Dropdown, Text
```

```python
import os

# Создание списка папок
folders = ['SUMMER', 'WINTER', 'SPRING']

# Базовый путь для создания папок
base_path = '/content'

# Перебор каждой папки из списка
for folder in folders:

    # Формирование пути к папке
    folder_path = os.path.join(base_path, folder)

    # Создание папки, если её нет
    os.makedirs(folder_path, exist_ok=True)

# Вывод сообщения об успешном создании папок
print("Folders created successfully!")
```

**ADD KEYWORDS**

```python
# Определение функции для добавления ключевых слов ко всем изображениям в папке
def add_keywords_to_folder(folder_path, new_keywords):
    try:
        for filename in os.listdir(folder_path):
            if filename.endswith(".jpeg") or filename.endswith(".jpg"):
                image_path = os.path.join(folder_path, filename)
                info = IPTCInfo(image_path)
                existing_keywords = info['keywords']
                if existing_keywords:
                    existing_keywords.append(new_keywords)
                    info['keywords'] = existing_keywords
                else:
                    info['keywords'] = [new_keywords]
                info.save()
                print(f"Keyword '{new_keywords}' added to {filename}")
        print("All images in the folder have been processed successfully!")
    except Exception as e:
        print(f"An error occurred: {e}")

# Определение функции для отображения существующих тегов
def display_existing_tags(folder_path):
    try:
        for filename in os.listdir(folder_path):
            if filename.endswith(".jpeg") or filename.endswith(".jpg"):
                image_path = os.path.join(folder_path, filename)
                info = IPTCInfo(image_path)
                existing_keywords = info['keywords']
                print(f"Метаданные для {filename}:")
                if existing_keywords:
                    print(f"Существующие ключевые слова: {existing_keywords}")
                else:
                    print("Существующие ключевые слова не найдены.")
    except Exception as e:
        print(f"Произошла ошибка: {e}")

# Определение функции для добавления ключевого слова к определенному изображению
def add_keyword_to_image(image_path, new_keyword):
    try:
        info = IPTCInfo(image_path)
        existing_keywords = info['keywords']
        if existing_keywords:
            existing_keywords.append(new_keyword)
            info['keywords'] = existing_keywords
        else:
            info['keywords'] = [new_keyword]
        info.save()
        print(f"Keyword '{new_keyword}' added to {image_path}")
    except Exception as e:
        print(f"Произошла ошибка: {e}")

# Определение функции для отображения существующих тегов и добавления ключевых слов
def display_and_add_keywords(folder_path, image_name, new_keyword):
    if image_name == "All Images":
        add_keywords_to_folder(folder_path, new_keyword)
    else:
        image_path = os.path.join(folder_path, image_name)
        add_keyword_to_image(image_path, new_keyword)

# Получение списка имен файлов изображений в папке
def get_image_filenames(folder_path):
    image_filenames = ["All Images"]
    for filename in os.listdir(folder_path):
        if filename.endswith(".jpeg") or filename.endswith(".jpg"):
            image_filenames.append(filename)
    return image_filenames

# Определение путей к папкам
folder_paths = ["/content/SUMMER", "/content/WINTER", "/content/SPRING"]

# Создание выпадающих меню для выбора папки и изображения
folder_dropdown = Dropdown(options=folder_paths, description='Папка:')
image_dropdown = Dropdown(options=["All Images"], description='Изображение:')

# Создание текстового поля для новых ключевых слов
keyword_input = Text(description='Новые ключевые слова:')

# Получение имен файлов изображений на основе выбранной папки
def update_image_dropdown(*args):
    selected_folder = folder_dropdown.value
    if selected_folder:
        image_filenames = get_image_filenames(selected_folder)
        image_dropdown.options = image_filenames

folder_dropdown.observe(update_image_dropdown, 'value')

# Создание интерактивного интерфейса
interact_manual(display_and_add_keywords, folder_path=folder_dropdown, image_name=image_dropdown, new_keyword=keyword_input)

def display_and_add_keywords(folder_path, image_name, new_keyword):
    """Display existing keywords for images in the selected folder and add a new keyword."""
```

**DELETE KEYWORDS**

```python
# Определение функции для отображения существующих ключевых слов для каждого изображения в папке
def display_existing_keywords(folder_path):
    try:
        for filename in os.listdir(folder_path):
            if filename.endswith(".jpeg") or filename.endswith(".jpg"):
                image_path = os.path.join(folder_path, filename)
                info = IPTCInfo(image_path)
                existing_keywords = [keyword.decode('utf-8') for keyword in info['keywords']]
                print(f"Метаданные для {filename}:")
                if existing_keywords:
                    print(f"Существующие ключевые слова: {existing_keywords}")
                else:
                    print("Ключевые слова не найдены.")
    except Exception as e:
        print(f"Произошла ошибка: {e}")

# Определение функции для удаления указанного ключевого слова из всех изображений в папке
def remove_keyword_from_images(folder_path, image_name, keyword_to_remove):
    try:
        if image_name == "All images":
            # Перебор всех изображений в папке
            for filename in os.listdir(folder_path):
                if filename.endswith(".jpeg") or filename.endswith(".jpg"):
                    image_path = os.path.join(folder_path, filename)
                    info = IPTCInfo(image_path)
                    existing_keywords = [keyword.decode('utf-8') for keyword in info['keywords']]
                    if keyword_to_remove in existing_keywords:
                        existing_keywords.remove(keyword_to_remove)
                        info['keywords'] = [keyword.encode('utf-8') for keyword in existing_keywords]
                        info.save()
                        print(f"Ключевое слово '{keyword_to_remove}' удалено из {filename}")
            print("Все изображения в папке успешно обработаны!")
        else:
            # Удаление указанного ключевого слова из выбранного изображения
            image_path = os.path.join(folder_path, image_name)
            info = IPTCInfo(image_path)
            existing_keywords = [keyword.decode('utf-8') for keyword in info['keywords']]
            if keyword_to_remove in existing_keywords:
                existing_keywords.remove(keyword_to_remove)
                info['keywords'] = [keyword.encode('utf-8') for keyword in existing_keywords]
                info.save()
                print(f"Ключевое слово '{keyword_to_remove}' удалено из {image_name}")
            else:
                print(f"Ключевое слово '{keyword_to_remove}' не найдено в {image_name}")
    except Exception as e:
        print(f"Произошла ошибка: {e}")

# Определение функции для обновления списка имен изображений на основе выбранной папки
def update_image_names(folder_path):
    try:
        image_names = ["All images"]
        for filename in os.listdir(folder_path):
            if filename.endswith(".jpeg") or filename.endswith(".jpg"):
                image_names.append(filename)
        return image_names
    except Exception as e:
        print(f"Произошла ошибка: {e}")
        return ["Все изображения"]

# Определение путей к папкам
folder_paths = ["/content/SUMMER", "/content/WINTER", "/content/SPRING"]

# Создание выпадающих списков для выбора папки, изображения и ввода ключевого слова для удаления
folder_dropdown = Dropdown(options=folder_paths, description='Папка:')
image_dropdown = Dropdown(options=["All images"], description='Изображение:')
keyword_input = Text(description='Ключевое слово для удаления:')

# Определение функции для отображения существующих ключевых слов и удаления указанного ключевого слова
def display_and_remove_keyword(folder_path, image_name, keyword_to_remove):
    display_existing_keywords(folder_path)
    remove_keyword_from_images(folder_path, image_name, keyword_to_remove)

# Обновление списка имен изображений при изменении выбранной папки
def update_image_dropdown(*args):
    selected_folder = folder_dropdown.value
    if selected_folder:
        image_dropdown.options = update_image_names(selected_folder)

folder_dropdown.observe(update_image_dropdown, 'value')

# Создание интерактивного интерфейса
interact_manual(display_and_remove_keyword, folder_path=folder_dropdown, image_name=image_dropdown, keyword_to_remove=keyword_input)

def display_and_add_keywords(folder_path, image_name, new_keyword):
    """Display existing keywords for images in the selected folder and add a new keyword."""
```

```python
import pandas as pd
from IPython.display import display, HTML

def display_image_metadata(folder_path):
    try:
        # Инициализация списков для хранения метаданных
        filenames = []
        metadata = []

        # Перебор всех файлов в папке
        for filename in os.listdir(folder_path):
            if filename.endswith(".jpg") or filename.endswith(".jpeg"):
                # Формирование полного пути к изображению
                image_path = os.path.join(folder_path, filename)
                # Получение метаданных для изображения
                info = IPTCInfo(image_path)
                # Сохранение имени файла
                filenames.append(filename)
                # Сохранение метаданных в виде словаря
                metadata.append(info)

        # Создание DataFrame из метаданных
        df = pd.DataFrame(metadata, index=filenames)

        # Добавление ярлыков к каждому полю
        columns = pd.MultiIndex.from_product([[folder_path], df.columns])

        # Присваивание MultiIndex столбцам DataFrame
        df.columns = columns

        # Стилизация DataFrame для цветного вывода
        styles = [
            dict(selector="th", props=[("background-color", "Orange"), ("color", "black")]),
            dict(selector="td", props=[("background-color", "lightgrey"), ("color", "black")]),
            dict(selector="tr:hover td", props=[("background-color", "skyblue"), ("color", "black")])
        ]
        styled_df = df.style.set_table_styles(styles)

        # Вывод стилизованного DataFrame
        display(HTML(styled_df.render()))

    except Exception as e:
        print(f"An error occurred: {e}")

# Определение путей к папкам
folders = ["SUMMER", "WINTER", "SPRING"]

# Перебор каждой папки и отображение метаданных изображений
for folder in folders:
    folder_path = os.path.join("/content", folder)
    print(f"Metadata for images in folder {folder}:")
    display_image_metadata(folder_path)
```
